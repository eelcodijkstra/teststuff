# Define pulse function and plot

```python
# Define a simple pulse function. Let's use 1/2 cycle of a cosine centered at 0, and zero everywhere else
def pulse(a):
    if a < np.pi/2.0 and a > -np.pi/2.0:
        return np.cos(a)
    else:
        return 0.0
```

```python
def pulse_usearrayinput(a):
    y = np.zeros(len(a))
    for i in range(0,len(a)): 
        y[i] = pulse(a[i])
    return y
```

```python
z = np.linspace(-5,5,100)
y = pulse_usearrayinput(z-1*1)
plt.plot(z,y)
plt.ylim(0,1.1)
```

```python
fig = plt.figure(figsize=(8,4))      # Create an empty figure
ax  = plt.axes()
line, = ax.plot([], [],linewidth=2)  # Create an empty line plot
plt.axis((0,1,-0.1,1.1))             # Set the bounds of the plot
#plt.xlim(-5,5)
ax.set_xlim(-5,5)
a = 1
t = np.linspace(0,5)
def plot_q(t):
    line.set_data(z,pulse_usearrayinput(z-a*t))  # Replace the line plot with the solution at time t
    
for t in range(0,5):
 time.sleep(1)
 ax.plot(z, pulse_usearrayinput(z-a*t))
 clear_output(wait=True)
 display(fig)
```

# Works - slow animation with clear_output(wait=True)

```python
fig = plt.figure(figsize=(8,4))      # Create an empty figure
ax  = plt.axes()
line, = ax.plot([], [],linewidth=2)  # Create an empty line plot
plt.axis((0,1,-0.1,1.1))             # Set the bounds of the plot
#plt.xlim(-5,5)
ax.set_xlim(-5,5)
txt = ax.text(0.1,0.8,'t = 0',transform = ax.transAxes)
a = 1
t = np.linspace(0,5)
def plot_q(t):
    line.set_data(z,pulse_usearrayinput(z-a*t))  # Replace the line plot with the solution at time t
    
for t in np.linspace(0,5,20):
    time.sleep(0.03)
    line.set_data(z,pulse_usearrayinput(z-a*t))  # Replace the line plot with the solution at time t
    txt.set_text('t = %.2f' % t)
    clear_output(wait=True)
    display(fig);

# close the figure at the end, so we don't get a duplicate of the last plot
plt.close()
```

```python
# From http://nbviewer.ipython.org/github/jakevdp/matplotlib_pydata2013/blob/master/notebooks/05_Animations.ipynb

import matplotlib.pyplot as plt
import numpy as np

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)
```

```python
# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
x = np.linspace(0, 2, 1000)

def animate(i):
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,
```

```python
from matplotlib import animation

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)
```
