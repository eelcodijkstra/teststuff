```python
import numpy as np
import matplotlib.pyplot as plt
from tempfile import NamedTemporaryFile
from IPython.display import HTML
from matplotlib import animation
```

# Try video with sinusoid

```python
# 12/23/14 - I added autoplay and loop to the video tag so
# it will automatically play and repeat
VIDEO_TAG = """<video controls autoplay loop>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=60, extra_args=['-vcodec', 'libx264', '-pix_fmt', 'yuv420p'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    return VIDEO_TAG.format(anim._encoded_video)

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))

numperiods = 2
numframes = 120
velocity = 1.0/numframes

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(0, numperiods), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    x = np.linspace(0, numperiods, 1000, endpoint=False)
    y = np.sin(2 * np.pi * (x - velocity * i))
    line.set_data(x, y)
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, 
                               animate, 
                               init_func=init,
                               frames=numframes, 
                               interval=20, 
                               blit=True
                               )

# call our new function to display the animation
display_animation(anim)
```

**Learned:**
- For a visually smooth animation the framerate should be high (50-60 fps, set in *anim_to_html*) and the difference in the graph from frame-to-frame should be quite small
- The FuncAnimation argument *interval* has no effect

# Try video with pulse

```python
# Define a simple arbitrary pulse function. Let's use 1/2 cycle of a cosine centered at 0,
# and zero everywhere else.

def pulse(a):
    if a < np.pi/2.0 and a > -np.pi/2.0:
        return np.cos(a)
    else:
        return 0.0
```

```python
# 12/23/14 - I added autoplay and loop to the video tag so
# it will automatically play and repeat
VIDEO_TAG = """<video controls autoplay loop>
 <source src="data:video/x-m4v;base64,{0}" type="video/mp4">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.mp4') as f:
            anim.save(f.name, fps=60, extra_args=['-vcodec', 'libx264', '-pix_fmt', 'yuv420p'])
            video = open(f.name, "rb").read()
        anim._encoded_video = video.encode("base64")
    return VIDEO_TAG.format(anim._encoded_video)

def display_animation(anim):
    plt.close(anim._fig)
    return HTML(anim_to_html(anim))

numframes = 180
zmin = -10
zmax = 10
numpnts = 200
u = float(zmax - zmin)/numframes

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(zmin, zmax), ylim=(-0.5, 2))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def plotpulse(t):
    z = np.linspace(zmin,zmax,numpnts)
    y = np.zeros(numpnts)
    for i in range(0,numpnts): 
        y[i] = pulse((z[i]-zmin) - u*t)
    line.set_data(z,y)
    return line,


# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, 
                               plotpulse, 
                               init_func=init,
                               frames=numframes, 
                               interval=20, 
                               blit=True
                               )

# call our new function to display the animation
display_animation(anim)
```
