```python
from matplotlib import animation
from JSAnimation import IPython_display

fig = plt.figure()
ax = plt.axes(xlim=(0, 10), ylim=(-2, 2))
line, = ax.plot([], [], lw=2)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(0, 10, 1000)
    y = np.sin(x - i * 0.02 * np.pi)
    line.set_data(x, y)
    return line,

animation.FuncAnimation(fig, animate, init_func=init,
                        frames=100, interval=30)
```

```python
# Define a simple arbitrary pulse function. Let's use 1/2 cycle of a cosine centered at 0,
# and zero everywhere else.

def pulse(a):
    if a < np.pi/2.0 and a > -np.pi/2.0:
        return np.cos(a)
    else:
        return 0.0
```

```python
numframes = 100
zmin = -10
zmax = 10
numpnts = 200
u = float(zmax - zmin)/numframes

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(zmin, zmax), ylim=(-0.5, 2))
line, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def plotpulse(t):
    z = np.linspace(zmin,zmax,numpnts)
    y = np.zeros(numpnts)
    for i in range(0,numpnts): 
        y[i] = pulse((z[i]-zmin) - u*t)
    line.set_data(z,y)
    return line,


# call the animator.  blit=True means only re-draw the parts that have changed.
animation.FuncAnimation(fig, 
                               plotpulse, 
                               init_func=init,
                               frames=numframes, 
                               interval=20, 
                               blit=True
                               )
```
