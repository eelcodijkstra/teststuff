# Built-in style sheets 

```python
print plt.style.available
```

```python
xval = np.linspace(0,4,100)
yval = np.sin(2*np.pi*xval)
```

## Default

```python
plt.plot(xval,yval)
```

## ggplot

```python
with plt.style.context(('ggplot')):
    plt.plot(xval,yval)
```

## dark_background

```python
with plt.style.context(('dark_background')):
    plt.plot(xval,yval)
```

## bmh

```python
with plt.style.context(('bmh')):
    plt.plot(xval,yval)
```

## grayscale

```python
with plt.style.context(('grayscale')):
    plt.plot(xval,yval)
```

## fivethirtyeight

```python
with plt.style.context(('fivethirtyeight')):
    plt.plot(xval,yval)
```
