```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.html.widgets import interact, fixed
```

# Set up some simple data to use in test plots

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

# Try some subplots from http://matplotlib.org/examples/pylab_examples/subplots_demo.html

```python
# Four axes, returned as a 2-d array
f, axarr = plt.subplots(2, 2)
axarr[0, 0].plot(x, y)
axarr[0, 0].set_title('Axis [0,0]')
axarr[0, 1].scatter(x, y)
axarr[0, 1].set_title('Axis [0,1]')
axarr[1, 0].plot(x, y ** 2)
axarr[1, 0].set_title('Axis [1,0]')
axarr[1, 1].scatter(x, y ** 2)
axarr[1, 1].set_title('Axis [1,1]')
# Fine-tune figure; hide x ticks for top plots and y ticks for right plots
plt.setp([a.get_xticklabels() for a in axarr[0, :]], visible=False)
plt.setp([a.get_yticklabels() for a in axarr[:, 1]], visible=False);
```

```python
# row and column sharing
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(x, y)
ax1.set_title('Sharing x per column, y per row')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r');
```

# Remove space between plots & adjust axis limits

```python
# row and column sharing
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
ax1.plot(x, y)
ax1.set_title('Sharing x per column, y per row')
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r');
# Set y limits
ax1.set_ylim(-1.2,1.2)
ax3.set_ylim(-1.2,1.2)
# Set x limits
ax3.set_xlim(-0.5,6.5)
ax4.set_xlim(-0.5,6.5)
# Fine-tune figure; make subplots close to each other and hide x ticks for
# all but bottom plot.
f.subplots_adjust(hspace=0,wspace=0)
#plt.setp([a.get_xticklabels() for a in f.axes[:-1]], visible=False);
```

# Change so use an Axes array

```python
# row and column sharing
f, ax = plt.subplots(2, 2, sharex='col', sharey='row')
ax[0][0].plot(x, y)
ax[0][0].set_title('Sharing x per column, y per row')
ax[0][1].scatter(x, y)
ax[1][0].scatter(x, 2 * y ** 2 - 1, color='r')
ax[1][1].plot(x, 2 * y ** 2 - 1, color='r');
# Set y limits
ax[0][0].set_ylim(-1.2,1.2)
ax[1][0].set_ylim(-1.2,1.2)
# Set x limits
ax[1][0].set_xlim(-0.5,6.5)
ax[1][1].set_xlim(-0.5,6.5)
# Fine-tune figure; make subplots close to each other.
f.subplots_adjust(hspace=0,wspace=0)
```

# Loop over Axes array to set plots & properties

```python
numrows = 2
numcols = 3
xlimitmin = -0.5
xlimitmax = 6.5
ylimit = 1.2
# row and column sharing
f, ax = plt.subplots(numrows, numcols, sharex='col', sharey='row',figsize=(10,8))
for i in range(numrows):
    for j in range(numcols):
        ax[i][j].plot(x,y)
ax[0][0].set_title('test title')
# Set y limits
for i in range(numrows):
    ax[i][0].set_ylim(-ylimit,ylimit)
    ax[i][0].set_ylabel('y')
# Set x limits
for j in range(numcols):
    ax[numrows-1][j].set_xlim(xlimitmin,xlimitmax)
    ax[numrows-1][j].set_xlabel('x')
# Fine-tune figure; make subplots close to each other.
f.subplots_adjust(hspace=0,wspace=0)
```

# Change dummy data to a circle & set aspect ratio to 1

```python
numpnts = 101
phi = np.linspace(0,2*np.pi,numpnts)
xx = np.cos(phi)
yy = np.sin(phi)
```

```python
numrows = 5
numcols = 5
xlimit = 1.2
ylimit = 1.2
# row and column sharing
f, ax = plt.subplots(numrows, numcols, sharex='col', sharey='row',figsize=(8,8))
for i in range(numrows):
    for j in range(numcols):
        ax[i][j].plot(xx,yy)
ax[0][0].set_title('test title')
# Set y limits
for i in range(numrows):
    ax[i][0].set_ylim(-ylimit,ylimit)
    ax[i][0].set_ylabel('y')
# Set x limits
for j in range(numcols):
    ax[numrows-1][j].set_xlim(-xlimit,xlimit)
    ax[numrows-1][j].set_xlabel('x')
# Fine-tune figure; make subplots close to each other.
f.subplots_adjust(hspace=0,wspace=0)
```
