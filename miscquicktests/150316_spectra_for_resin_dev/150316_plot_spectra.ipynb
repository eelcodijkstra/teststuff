```python
from __future__ import division
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
BLSabsorption = np.loadtxt('BLS_in_ethyl_acetate.csv', delimiter=',')
UV326absorption = np.loadtxt('UV326absorptionspectrum.csv', delimiter=',')
Irgacure819absorption = np.loadtxt('Irgacure819absorptionspectrum.csv', delimiter=',')
TMDPOabsorption = np.loadtxt('TMDPOabsorptionspectrum.csv', delimiter=',')
LED385nm_Thorlabs = np.loadtxt('385nm_LED_Thorlabs_M385L2.csv', delimiter=',')
```

```python
print('BLS array shape:',np.shape(BLSabsorption))
print('UV326 array shape:',np.shape(UV326absorption))
print('Irgacure array shape:',np.shape(Irgacure819absorption))
print('TMDPO array shape:',np.shape(TMDPOabsorption))
print('385 nm LED array shape:',np.shape(LED385nm_Thorlabs))
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_title('Current Resin',fontsize=18)
ax.plot(LED385nm_Thorlabs[:,0],LED385nm_Thorlabs[:,1], label='Thorlabs 385 nm LED')
ax.plot(BLSabsorption[:,0],BLSabsorption[:,1], label='BLS-99 UV absorber', color='k')
ax.plot(TMDPOabsorption[:,0],TMDPOabsorption[:,1], label='TMDPO photoinitiator', color='r')
ax.legend(bbox_to_anchor=(1.2, 0.9))
ax.set_xlabel('Wavelength (nm)',fontsize=12)
ax.set_ylabel('Absorption/Normalized Emission',fontsize=12)
ax.set_ylim(0,1.2)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_title('Replace TMDPO With Irgacure 819',fontsize=18)
ax.plot(LED385nm_Thorlabs[:,0],LED385nm_Thorlabs[:,1], label='Thorlabs 385 nm LED')
ax.plot(BLSabsorption[:,0],BLSabsorption[:,1], label='BLS-99 UV absorber', color='k')
ax.plot(Irgacure819absorption[:,0],Irgacure819absorption[:,1], label='Irgacure 819', color='r')
ax.legend(bbox_to_anchor=(1.2, 0.9))
ax.set_xlabel('Wavelength (nm)',fontsize=12)
ax.set_ylabel('Absorption/Normalized Emission',fontsize=12)
ax.set_ylim(0,1.2)
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.set_title('Replace BLS-99 With UV-326',fontsize=18)
ax.plot(LED385nm_Thorlabs[:,0],LED385nm_Thorlabs[:,1], label='Thorlabs 385 nm LED')
ax.plot(UV326absorption[:,0],UV326absorption[:,1], label='UV-326 UV absorber', color='k')
ax.plot(Irgacure819absorption[:,0],Irgacure819absorption[:,1], label='Irgacure 819', color='r')
ax.legend(bbox_to_anchor=(1.2, 0.9))
ax.set_xlabel('Wavelength (nm)',fontsize=12)
ax.set_ylabel('Absorption/Normalized Emission',fontsize=12)
ax.set_ylim(0,1.2)
```
