```python
from ctypes import *
from dwfconstants import *
import math
import time
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
import sys
import numpy as np
from scipy.signal import square
import collections

if sys.platform.startswith("win"): #windows
    dwf = cdll.dwf
elif sys.platform.startswith("darwin"): #MacOS
    dwf = cdll.LoadLibrary("libdwf.dylib")
else:
    dwf = cdll.LoadLibrary("libdwf.so") #Linux 
```

```python
#declare ctype variables
hdwf = c_int()
InChannel0 = c_int(0)  #PMT
InChannel1 = c_int(1)  #Applied on sample
OutChannel0 = c_int(0)
OurChannel1 = c_int(1)

#print DWF version
version = create_string_buffer(16)
dwf.FDwfGetVersion(version)
print "DWF Version: "+version.value
#Just in case attached device wasn't closed by previous use
dwf.FDwfDeviceCloseAll()
#open device
print "Opening first device..."
dwf.FDwfDeviceOpen(c_int(-1), byref(hdwf))

if hdwf.value == hdwfNone.value:
    print "failed to open device"
    quit()

    
#------------------------- Metal -------------------------#

metal = str(raw_input(r'Input the name of metal: '))    

looprange = range(8)
Vpp_set = 7
    
#------------------------- Wave Parameters -------------------------#

DataBufferSize = c_int(8192)
Nloop = c_int(3)
VoutFreqArray = np.array([1e6, 3e5, 1e5, 3e4, 1e4, 3e3, 1e3, 500])
TdelayMicrosArray = np.array([10, 10, 10, 16.67, 50, 166.67, 500, 1000])
TsampleMicrosArray = np.array([0.01, 0.01, 0.01, 0.01, 0.02, 0.06, 0.2, 0.4])
saveData_0 = np.zeros((len(VoutFreqArray), DataBufferSize.value, Nloop.value))
saveData_1 = np.zeros((len(VoutFreqArray), DataBufferSize.value, Nloop.value))
timeAxis_us = np.zeros((DataBufferSize.value, len(VoutFreqArray)))

for n in looprange:
    #WaveformIndex = c_byte(input("WaveformIndex: "))
    #funcDC = 0, funcSine = 1, funcSquare = 2, funcTriangle = 3, funcRampUp   = 4, 
    #funcRampDown = 5, funcNoise = 6, funcCustom = 30, funcPlay = 31
    #VoutFreq = c_double(input("Frequency: "))
    #Nperiods = c_double(input("Number of Periods: "))
    #Vpp = c_double(input("Vpp: "))
    #Offset = c_double(input("Offset: "))
    #Phase_Deg = c_double(input("Phase_deg: "))
    #TdelayMicros = c_double(input("Delay (us): "))
    #TsampleMicros = c_double(input("Sampling time (us): "))

    WaveformIndex = c_byte(2)
    #funcDC = 0, funcSine = 1, funcSquare = 2, funcTriangle = 3, funcRampUp   = 4, 
    #funcRampDown = 5, funcNoise = 6, funcCustom = 30, funcPlay = 31
    
    VoutFreq = c_double(VoutFreqArray[n]) ####
    Nperiods = c_double(0.5)
    Vpp = c_double(Vpp_set) ###?
    Offset = c_double(0)
    Phase_Deg = c_double(0)
    TdelayMicros = c_double(TdelayMicrosArray[n]) ####
    TsampleMicros = c_double(TsampleMicrosArray[n]) ####
    InCh0range = c_double(5)
    InCh1range = c_double(50)
    
    print "------------------------- " + str(VoutFreq.value) + " Hz -------------------------"
    
    
    
#------------------------- Set up Analog Out -------------------------#

    dwf.FDwfAnalogOutNodeEnableSet(hdwf, OutChannel0, AnalogOutNodeCarrier, c_bool(True)) #Enable channel 0
    dwf.FDwfAnalogOutNodeFunctionSet(hdwf, OutChannel0, AnalogOutNodeCarrier, WaveformIndex) #Set Waveform
    dwf.FDwfAnalogOutNodeFrequencySet(hdwf, OutChannel0, AnalogOutNodeCarrier, VoutFreq) #Set frequency
    OutFreqGet = c_double()
    dwf.FDwfAnalogOutNodeFrequencyGet(hdwf, OutChannel0, AnalogOutNodeCarrier, byref(OutFreqGet))
    print "Output Frequency: " + str(OutFreqGet.value) + " Hz"
    dwf.FDwfAnalogOutNodeAmplitudeSet(hdwf, OutChannel0, AnalogOutNodeCarrier, c_double(Vpp.value/2)) #Set Amplitude
    OutAmpGet = c_double()
    dwf.FDwfAnalogOutNodeAmplitudeGet(hdwf, OutChannel0, AnalogOutNodeCarrier, byref(OutAmpGet))
    print "Output Vpp: " + str(OutAmpGet.value*2) + " V"
    dwf.FDwfAnalogOutNodeOffsetSet(hdwf, OutChannel0, AnalogOutNodeCarrier, Offset) #Set Offset
    OutOffsetGet = c_double()
    dwf.FDwfAnalogOutNodeOffsetGet(hdwf, OutChannel0, AnalogOutNodeCarrier, byref(OutOffsetGet))
    print "Output Offset: " + str(OutOffsetGet.value) + " V"
    dwf.FDwfAnalogOutNodePhaseSet(hdwf, OutChannel0, AnalogOutNodeCarrier, Phase_Deg) #Set Phase
    OutPhaseGet = c_double()
    dwf.FDwfAnalogOutNodePhaseGet(hdwf, OutChannel0, AnalogOutNodeCarrier, byref(OutPhaseGet))
    print "Output Phase: " + str(OutPhaseGet.value) + " deg"
    pulseDuration_seconds = c_double(Nperiods.value/VoutFreq.value)
    dwf.FDwfAnalogOutRunSet(hdwf, OutChannel0, pulseDuration_seconds) #Set pulse length
    OutRunGet = c_double()
    dwf.FDwfAnalogOutRunGet(hdwf, OutChannel0, byref(OutRunGet))
    print "Pulse length: " + str(OutRunGet.value*1e3) + " ms"
    dwf.FDwfAnalogOutWaitSet(hdwf, OutChannel0, c_double(TdelayMicros.value*1e-6)) #set how long to wait after trigger to start waveform
    OutWaitGet = c_double()
    dwf.FDwfAnalogOutWaitGet(hdwf, OutChannel0, byref(OutWaitGet))
    print "Wait time after trigger: " + str(OutWaitGet.value*1e3) + " ms"
    dwf.FDwfAnalogOutTriggerSourceSet(hdwf, OutChannel0, c_int(1)) #trigger source: 1=PC, 2=AnalogIn Detector
    dwf.FDwfAnalogOutRepeatTriggerSet(hdwf, OutChannel0, c_bool(True)) #wait for trigger for each repeat
    dwf.FDwfAnalogOutRepeatSet(hdwf, OutChannel0, Nloop) #set repeat count to Nloop
    
    
    
#------------------------- Set up Analog In -------------------------#

    dwf.FDwfAnalogInReset(hdwf) #Reset so we start from a known state
    sampling_frequency = c_double(1/(TsampleMicros.value*1e-6))
    dwf.FDwfAnalogInFrequencySet(hdwf, sampling_frequency)
    samplingFreqGet = c_double()
    dwf.FDwfAnalogInFrequencyGet(hdwf, byref(samplingFreqGet))
    print "Sampling Frequency: " + str(samplingFreqGet.value/1e6) + " MHz"
    #Set which filtering to use on the channel
    dwf.FDwfAnalogInChannelFilterSet(hdwf, c_int(-1), c_int(1)) #Choose between 0=decimate, 1=average, 2=min/max
    ###### ? Get a single buffer full of data for each trigger event
    dwf.FDwfAnalogInAcquisitionModeSet(hdwf, c_int(0)) #Single=0, ScanShift=1, ScanScreen=2, Record=3
    #Make data buffer in which to put acquired data
    BufferFillTime_us = c_double(DataBufferSize.value * TsampleMicros.value) #calculate how long it should take to fill buffer
    print "Time to fill buffer: " + str(BufferFillTime_us.value) + " us"
    nBits = c_int()
    dwf.FDwfAnalogInBitsInfo(hdwf, byref(nBits)) #How many ADC bits?
    print "Number of ADC bits: " + str(nBits.value)
#    DataBufferSize = c_int(8192)
    dwf.FDwfAnalogInTriggerAutoTimeoutSet(hdwf, c_int(0)) #disable auto trigger
    dwf.FDwfAnalogInTriggerSourceSet(hdwf, c_int(1)) #trigger source = 1 is trigger from PC
    dwf.FDwfAnalogInTriggerPositionSet(hdwf, c_double(BufferFillTime_us.value/1e6/2)) #set trigger position to beginning of data buffer
    voltsInMin = c_double()
    voltsInMax = c_double()
    nSteps = c_double()
    dwf.FDwfAnalogInChannelRangeInfo(hdwf, byref(voltsInMin), byref(voltsInMax), byref(nSteps)) #get channel range info
    print "Input Min range: " + str(voltsInMin.value) + " V"
    print "Input Max range: " + str(voltsInMax.value) + " V"
    print "nSteps range: " + str(nSteps.value)
    dwf.FDwfAnalogInChannelRangeSet(hdwf, InChannel0, InCh0range) #set voltage range for channel 0
    InCh0rangeGet = c_double()
    dwf.FDwfAnalogInChannelRangeGet(hdwf, InChannel0, byref(InCh0rangeGet)) #get current voltage range for channel 0
    print "Final range Channel 0: " + str(InCh0rangeGet.value) + " V"
    dwf.FDwfAnalogInChannelRangeSet(hdwf, InChannel1, InCh1range) #set voltage range for channel 1
    InCh1rangeGet = c_double()
    dwf.FDwfAnalogInChannelRangeGet(hdwf, InChannel1, byref(InCh1rangeGet)) #get current voltage range for channel 1
    print "Final range Channel 1: " + str(InCh1rangeGet.value) + " V"

    time.sleep(2) #wait 2 sec for everything to stabilize
    
    
    
#------------------------- Start Analog Out and In -------------------------#

    ###    % Make time interval array to use for x-axis
    timeAxis_us[ : , n] = np.arange(0, DataBufferSize.value*TsampleMicros.value, TsampleMicros.value)

    #Start Analog Out and Analog In (they should wait until recieving a trigger before they fire)
    print "Turn on Analog Out & In"
    #dwf.FDwfAnalogInConfigure(hdwf, InChannel0, c_bool(True))
    #dwf.FDwfAnalogOutConfigure(hdwf, OutChannel0, c_bool(True))

    c_data_0 = (c_double*DataBufferSize.value)()
    c_data_1 = (c_double*DataBufferSize.value)()
    #saveData_0 = np.zeros((DataBufferSize.value, Nloop.value))
    #saveData_1 = np.zeros((DataBufferSize.value, Nloop.value))
    STS = c_byte()


    for i in range(Nloop.value):
        dwf.FDwfAnalogOutConfigure(hdwf, OutChannel0, c_bool(True))
        dwf.FDwfAnalogInConfigure(hdwf, InChannel0, c_bool(True))
        time.sleep(0.005)
        dwf.FDwfDeviceTriggerPC(hdwf) #Send trigger from PC
        time.sleep((BufferFillTime_us.value + 0.5) * 1e-3) #% Pause for xx milliseconds to fill Analog In buffer in Analog Discovery
        dwf.FDwfAnalogInStatus(hdwf, c_bool(True), byref(STS))
        dwf.FDwfAnalogInStatusData(hdwf, InChannel0, byref(c_data_0), DataBufferSize) #read data from buffer in DWF
        dwf.FDwfAnalogInStatusData(hdwf, InChannel1, byref(c_data_1), DataBufferSize)
        for j in range(DataBufferSize.value): #put data in arrays for plotting
            saveData_0[n, j, i] = c_data_0[j]
            saveData_1[n, j, i] = c_data_1[j]
        print "Loop iteration: " + str(i)
        dwf.FDwfAnalogOutConfigure(hdwf, OutChannel0, c_bool(False))
        dwf.FDwfAnalogInConfigure(hdwf, InChannel0, c_bool(False))
        time.sleep(15) #Pause for xx seconds
        
#Clean up
dwf.FDwfDeviceCloseAll()



#------------------------- Write data to file & Plot -------------------------#

for n in looprange:
    WaveformIndex = c_byte(2)
    #funcDC = 0, funcSine = 1, funcSquare = 2, funcTriangle = 3, funcRampUp   = 4, 
    #funcRampDown = 5, funcNoise = 6, funcCustom = 30, funcPlay = 31
    Nloop = c_int(3)
    VoutFreq = c_double(VoutFreqArray[n]) ####
    Nperiods = c_double(0.5)
    Vpp = c_double(Vpp_set)
    Offset = c_double(0)
    Phase_Deg = c_double(0)
    TdelayMicros = c_double(TdelayMicrosArray[n]) ####
    TsampleMicros = c_double(TsampleMicrosArray[n]) ####
    InCh0range = c_double(5)
    InCh1range = c_double(50)
    
    pulseDuration_seconds = c_double(Nperiods.value/VoutFreq.value)
    BufferFillTime_us = c_double(DataBufferSize.value * TsampleMicros.value)
    
    for i in range(Nloop.value):
        xData = timeAxis_us[:, n]
        yChan0 = saveData_0[n, :, i]
        yChan1 = saveData_1[n, :, i]

        #--------------- Write to file ---------------#
        np.savetxt("dat_" + metal + "_" + str("%.0f" %VoutFreq.value)+ "Hz_" + str(i+1) + ".txt", 
                   np.array([xData, yChan0, yChan1]).T, delimiter='\t', header='time_us\tPMT_volts\tVpp_volts')


        #--------------- Plot ---------------#
        fig, ax0 = plt.subplots(figsize=(15, 8))
        ax1 = ax0.twinx()
        ax0.plot( xData, yChan0, 'r-')
        ax1.plot( xData, yChan1, 'b--')
        ax0.set_xlim(0, DataBufferSize.value*TsampleMicros.value)
        #ax0.set_title(str(VoutFreq.value) + " Hz", fontsize=18)
        ax0.set_xlabel('Time ($\mu$s)', fontsize=18)
        ax0.set_ylabel('PMT Output (V)', color='r', fontsize=18)
        ax0.tick_params(axis='x', labelsize=16)
        ax0.tick_params(axis='y', labelsize=16)
        #ax0.text(1, 1, str(VoutFreq.value) + " Hz\n" + str(Nperiods.value) + " T", ha='right', va='top')
        #ax0.spines['left'].set_color('red')
        ax1.set_ylabel('Voltage Applied to Sample (V)', color='b', fontsize=18)
        ax1.tick_params(axis='y', labelsize=16)
        #ax1.spines['right'].set_color('blue')

        #box dimensions in axes coords
        left, width = 1.1, .3
        bottom, height = 0, 1
        right = left + width
        top = bottom + height

        # axes coordinates are 0,0 is bottom left and 1,1 is upper right
        p = patches.Rectangle(
            (left, bottom), width, height,
            fill=False, transform=ax0.transAxes, clip_on=False
            )

        ax0.add_patch(p)

        if WaveformIndex.value is 1:
            waveform = "Sine"
        else:
            waveform = "Square"

        #content in box
        parametersDWF = "Experiment Parameters: \n" \
                    + "Waveform: " + waveform + "\n" \
                    + "Frequency: " + str(VoutFreq.value) + " Hz\n" \
                    + "Func. Gen. Vpp: " + str(Vpp.value) + " V\n" \
                    + "Offset: " + str(Offset.value) + " V\n" \
                    + "Phase: " + str(Phase_Deg.value) + "$^\circ$\n" \
                    + "Pulse Length: " + str("%.3f" %(pulseDuration_seconds.value*1e3)) + " ms\n" \
                    + "Number of Periods: " + str(Nperiods.value) + "\n" \
                    + "Delay: " + str(TdelayMicros.value) + " us\n" \
                    + "Sampling Period: " + str(TsampleMicros.value) + " us\n" \
                    + "Buffer Time: " + str("%.3f" %(BufferFillTime_us.value)) + " us\n" \
                    + "In Ch0 Range: " + str(InCh0range.value) + " V\n" \
                    + "In Ch1 Range: " + str(InCh1range.value) + " V\n" \
                    + "-----------------------\n" \
                    + "Measured Data: \n" \
                    + "Ch0Min: " + str("%.3f" %np.amin(yChan0)) + " V\n" \
                    + "Ch0Max: " + str("%.3f" %np.amax(yChan0)) + " V\n" \
                    + "Ch0Min-Ch0Min: " + str("%.3f" %(np.amax(yChan0)-np.amin(yChan0))) + " V\n" \
                    + "Ch1Min: " + str("%.3f" %np.amin(yChan1)) + " V\n" \
                    + "Ch1Max: " + str("%.3f" %np.amax(yChan1)) + " V\n" \
                    + "Ch1Min-Ch1Min: " + str("%.3f" %(np.amax(yChan1)-np.amin(yChan1))) + " V\n" \

        #set text in box
        ax0.text(left+0.01, top-0.01, parametersDWF,
                horizontalalignment='left',
                verticalalignment='top',
                transform=ax0.transAxes, fontsize=18)
        
        ax0.text(0.95, 0.95, metal,
                horizontalalignment='right',
                verticalalignment='top',
                transform=ax0.transAxes, fontsize=32)
        

        plt.savefig("G_" + metal + "_" + str("%.0f" %VoutFreq.value)+ "Hz_" + str(i+1) + ".png", bbox_inches='tight') #save figure, bbox_inches=tight:save all content in figure
        plt.show()
```
